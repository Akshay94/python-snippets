```python
import pandas as pd
import pandas_datareader.data as web
```

```python
df = web.DataReader('AAPL', 'iex', '2017-01-01', '2017-12-31')
```

```python
df.index = pd.to_datetime(df.index)
```

```python
print(df.shape)
```

```python
print(df.head())
```

```python
print(df.tail())
```

```python
d_ohlc = {'open': 'first',
          'high': 'max',
          'low': 'min',
          'close': 'last'}
```

```python
print(df.resample('MS').agg(d_ohlc))
```

```python
print(df.resample('QS').agg(d_ohlc))
```

```python
print(df.resample('2W-MON', closed='left', label='left').agg(d_ohlc))
```

```python
d_ohlcv = {'open': 'first',
           'high': 'max',
           'low': 'min',
           'close': 'last',
           'volume': 'sum'}
```

```python
print(df.resample('MS').agg(d_ohlcv))
```
